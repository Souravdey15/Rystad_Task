```python
data['Date'] = pd.to_datetime(data['Date'])
data_by_year=[]

result = data.groupby(data.Date.dt.to_period("Y"))['Total US Refinery Net Input of Crude Oil'].sum()
result_2016_padd1 = data.groupby(data.Date.dt.to_period("Y"))['(PADD 1) Refinery and Blender Net Input of Crude Oil'].sum()
result_2016_padd2 = data.groupby(data.Date.dt.to_period("Y"))['(PADD 2) Refinery and Blender Net Input of Crude Oil'].sum()
result_2016_padd3 = data.groupby(data.Date.dt.to_period("Y"))['(PADD 3) Refinery and Blender Net Input of Crude Oil'].sum()
result_2016_padd4 = data.groupby(data.Date.dt.to_period("Y"))['(PADD 4) Refinery and Blender Net Input of Crude Oil'].sum()
result_2016_padd5 = data.groupby(data.Date.dt.to_period("Y"))['(PADD 5) Refinery and Blender Net Input of Crude Oil'].sum()

for year, records in result_2016_padd1.items():
    data_by_year.append(f"(PADD 1) Refinery and Blender Net Input of Crude Oil {year} = {records}")
for year, records in result_2016_padd2.items():
    data_by_year.append(f"(PADD 2) Refinery and Blender Net Input of Crude Oil {year} = {records}")
for year, records in result_2016_padd3.items():
    data_by_year.append(f"(PADD 3) Refinery and Blender Net Input of Crude Oil {year} = {records}")
for year, records in result_2016_padd4.items():
    data_by_year.append(f"(PADD 4) Refinery and Blender Net Input of Crude Oil {year} = {records}")
for year, records in result_2016_padd5.items():
    data_by_year.append(f"(PADD 5) Refinery and Blender Net Input of Crude Oil {year} = {records}")
for year, records in result.items():    
    data_by_year.append(f"Total US Refinery Net Input of Crude Oil {year} = {records}")
    
```
